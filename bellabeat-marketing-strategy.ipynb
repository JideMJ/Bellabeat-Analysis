### Data Analysis on Sleep.
To get the most on the analysis on sleep patterns and quality, the dataframe is adjusted to reflect the quality of sleep to fit into three categories according to WHO guidelines for adults; 

Poor sleep is below 360 minutes (6 hours), 

adequate sleep is between 360 minutes and 480 minutes (6-8 hours), 

while oversleep category is 480 minutes (8 hours) and above. 

```r
daily_sleep_v1 <- daily_sleep%>%
  mutate(sleep_quality = case_when(TotalMinutesAsleep<= 360 ~ "poor_sleep", 
                                TotalMinutesAsleep>= 360 & TotalMinutesAsleep<= 480 ~ "adequate_sleep", 
                                TotalMinutesAsleep> 480 ~ "oversleep", TRUE ~ "poor_sleep"))
```

The analysis also covers sleep latency, which is also called sleep onset latency and defined as the amount of time it takes to fall asleep after the lights have been turned off. Ideally, it should take a brief amount of time to fall asleep after the lights have been turned off, but not too long. 
According to the Journal of Clinical Sleep Medicine, sleep latency is related directly to sleep efficiency. If an individual spends the majority of the time they are in bed during an evening actually asleep, then they are considered to be sleep efficient. However, if an individual spends a lot of the time they are in bed not actually asleep, but trying to sleep, then that is not considered to be sleep efficient.
Sleep latency relates directly to sleep efficiency, because if a person is able to fall asleep quickly, they are more likely to have an efficient sleep. If a person has ideal sleep latency, they are likely to have good sleep efficiency as well, since the two go hand in hand.
Sleep latency varies from person to person, as it does in the analysis. 

To examine sleep latency in the analysis, we check how quickly the users are able to fall asleep from the total time spent in bed:

```r
daily_sleep_v1$SleepLatency = daily_sleep_v1$TotalTimeInBed - daily_sleep_v1$TotalMinutesAsleep
```

Normal adult mean sleep latency is between 10 and 20 min. According to the American Academy of Sleep Medicine (AASM), a sleep latency of <8 min is diagnostic of sleepiness, defined for this analysis as quick latency. Conversely, sleep latency greater than 20 minutes is deemed slow latency, that is, the individual is unable to fall asleep as promptly as they ought to. 

```r
daily_sleep_v2 <- daily_sleep_v1%>%
  mutate(SleepOnset = case_when(SleepLatency<= 10 ~ "quick_latency",
                                SleepLatency>= 10 & SleepLatency<=20 ~ "normal_latency",
                                SleepLatency>= 20 ~ "slow_latency", TRUE ~ "quick_latency"))
```

Majority of Bellabeat users get adequate sleep, within 6 to 8 hours.

### Resting Heart rate. 
Heart rate, or pulse, is the number of times the heart beats in 1 minute. Heart rates vary from person to person. It’s lower when an individual is at rest and higher during exercise. As Bellabeat's devices tracks heart rates in users, it can be helpful in knowing how to figure out each person's best exercise program. 
According to American Heart Association (AHA), a normal resting heart rate for adults ranges from 60 to 100 beats per minute. Generally, a lower heart rate at rest implies more efficient heart function and better cardiovascular fitness.

Heart rate among Bellabeat Users are herein collated and examined according to metric from AHA. Hence, normal heart rate is reported as a range from 60 to 100 beats per minute in value. Individuals with heart rates above 100 are recorded with a "fast" heart rate, while "slow" heart rate is below 60 beats per minute. 

```r
heartrate_log <- heartrate_log %>%
  mutate(heart_rate = case_when(Value>= 60 & Value <= 100 ~ "Normal", 
                                Value<= 59 ~ "Slow", 
                                Value> 100 ~ "Quick", TRUE ~ "Slow"))
```

##### Proportion of Bellabeat Users and heart rates. 
Ratio of users with normal, slow and quick heart rates are plotted for graphical representation. 

```r
ggplot(data = heartrate_log) +
  geom_bar(mapping = aes(x= heart_rate, fill=heart_rate)) +
  labs(title = "Bellabeat Users' Resting Heart rate.")
```

Most users have the heart rate around the normal range. Users with the slow heart rate could very well be highly trained heart rates. Research conducted for professional athletes have indicated a normal resting heart rate closer to 40 beats per minute. Understanding of the report above should take this to account.
